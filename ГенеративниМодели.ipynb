# Генеративни модели

## Општи концепти на генеративни модели

Генеративните модели во машинско учење се базираат на Баесовиот концепт на учење во кој веројатноста за некоја хипотеза се базира на тоа 

Баесовото учење се базира на Баесовото правило со кое може да се изрази условната веројатност да се случи еден настан $Y$ (хипотеза) доколку се случил друг настан $X$ (доказ) и гласи
\begin{equation}
p(Y|X)=\frac{p(Y)p(X|Y)}{p(X)},
\end{equation}
каде што веројатноста $p(Y)$ се нарекува приорна веројатност за $Y$, $p(X|Y)$ се нарекува веродостојност (likelihood), а $p(X)$ се нарекува маргинална веродостојност.

(следнава равенка засега не е потребна)
\begin{equation}
p(Y=y|X=x)=\frac{P(Y=y,X=x)}{p(X=x)}=\frac{p(Y=y)p(X=x|Y=y)}{\sum_{y'}p(Y=y')p(X=x|Y=y')}
\end{equation}

Ако се употреби Баесовото правило може да се изрази општа форма на генеративен класификатор
\begin{equation}
p(y=c|\mathbf{x},\theta)=\frac{p(y=c|\theta)p(\mathbf{x}|y=c,\theta)}{\sum_{c'}p(y=c'|\theta)p(\mathbf{x}|y=c',\theta)},
\end{equation}
каде што $y$ е целната променлива, $c$ е конкретна вредност (класа), $\mathbf{x}$ се атрибутите на даден примерок, и со $\theta$ се претставени параметрите на моделот.



Во ова поглавје подетално ќе бидат разгледани неколку едноставни генеративни методи како што се Гаусов наивен Баесов класификатор, Квадратна дискриминативна анализа, и Линеарна дискриминативна анализа. Истите ќе бидат испробани за решавање на проблем на класификација и намалување на димензионалност во Ирис податочното множество.


## Наивен Баесов класификатор

Еден од наједноставните генеративни модели е наивниот Баесов класификатор во кој се претпоставува дека влезните атрибути се условно независни. Доколку бројот на атрибути го означиме со $D$ веродостојноста може да се изрази како
\begin{equation}
p(\mathbf{x}|y=c,\theta)=\prod_{j=1}^D p(x_j|y=c,\theta_{jc}).
\end{equation}

Постојат повеќе варијанти на овој модел, но тука ќе го разгледаме само Гаусовиот наивен Баесов класификатор кој работи со реални вредности.

### Гаусов наивен Баесов класификатор

Доколку се претпостави дека условните веројатности за вредностите на атрибутите следат нормална дистрибуција
\begin{equation}
p(x_j|y=c,\theta_{jc})=\mathcal{N}(\mathbf{x}|\mu_{jc},\sigma_{jc}^2),
\end{equation}
каде што $\mu_{jc}$ и $\sigma_{jc}^2$ се средната вредност и варијансата на вредностите на атрибутот $j$ за класата $c$.
Вака дефинираниот модел се нарекува Гаусов наивен Баесов класификатор.

## Гаусова дискриминативна анализа 

Во оваа група на модели се користи повеќедимензионална Гаусова дистрибуција за дефинирање на класните условни веројатности (веродостојностите), па така во генеративен класификатор со форма
\begin{equation}
p(y=c|\mathbf{x},\theta) \propto p(y=c|\theta)p(\mathbf{x}|y=c,\theta)
\end{equation}
за секоја класа $c$ ќе имаме
\begin{equation}
p(\mathbf{x}|y=c,\theta)=\mathcal{N}(\mathbf{x}|\mu_c,\Sigma_c),
\end{equation}
каде што $\mu_c$ е вектор од средните вредности на атрибутите за податоците од класа $c$, a $\Sigma_c$ е коваријансна матрица за податоците од класа $c$.

Алтернативно, ако се пресмета логаритам од двете страни се добива
\begin{equation}
\log(p(y=c|\mathbf{x},\theta)) \propto \log(p(y=c|\theta)) + \log(p(\mathbf{x}|y=c,\theta)).
\end{equation}

Дополнително треба да се зададат или пресметаат вредностите за приорите за секоја класа $p(y=c|\theta)$, кои може во некои случаи да се исти, а во други да се различни.

На крај, како предвидена класа се одбира таа што има најголема постериор веројатност.

Зависно од карактеристиките на матриците $\Sigma_c$ постојат повеќе типови на гаусови дискриминативни анализи кои ќе бидат разгледани во следните подпоглавја.

### Квадратна дискриминативна анализа

Во квадратната дискриминативна анализа (Quadratic discriminant analysis - QDA) се користи целосната коваријансна матрица, што резултира со постоење на квадратни членови од $\mathbf{x}$ во изразите за условната веројатност. На тој начин со овој метод може да се изразат и граници на одлука помеѓу класите во форма на квадратна функција. Воедно вкупниот број на параметри во QDA е релативно голем. Поради сето ова може да се каже дека QDA e доста флексибилна и експресивна.

Доколку коваријансните матрици за сите класи се дијагонални, моделот станува еквивалентен со наивен Баесов класификатор.

### Линеарна дискриминативна анализа

Во линеарната дискриминативна анализа (Linear discriminant analysis - LDA) коваријансната матрица $\Sigma_c$ е иста за сите класи. Поради тоа доаѓа до поништување на квадратните членови од $\mathbf{x}$ со што границите на одлука помеѓу класите добиваат форма на линеарна функција. Поради споделувањето на коваријансната матрица помеѓу класите овој метод има значително помалку параметри од QDA. Од една страна тоа го прави помалку флексибилен, но од друга страна со тоа станува помалку склон кон пренагодување.

Доколку коваријансната матрица $\Sigma_c$ освен што е споделена помеѓу класите има и дијагонална форма, тогаш станува збор за дијагонална линеарна дискриминативна анализа (Diagonal linear discriminant analysis - DLDA)

### Вчитување на податочното множество

Ирис податочното множество е вклучено во Scikit-learn библиотеката, па можеме да го вчитаме од неа со едноставен функциски повик. Потоа ги вчитуваме во посебни променливи X и y, атрибутите и целната променлива. Исто така можеме да ги вчитаме и имињата на класите од целната променлива, како и имињата на атрибутите.

```python
iris = datasets.load_iris()

X = iris.data # Матрица која ги содржи атрибутите на примероците од податочното множество. Секој ред соодвествува со една инстанца, а секоја колона соодветсвува на еден атрибут.
y = iris.target # Вектор од вредностите на целната променлива за секоја инстанца од податочното множество.

target_names = iris.target_names # листа со различните вредности (класи) на целната променлива
feature_names = iris.feature_names # листа со имињата на атрибутите

number_of_classes = len(target_names) # Број на различни класи. Може да се најде и преку np.unique(y)
number_of_features = len(feature_names)  # Број на атрибути. Може да се најде и преку X.shape[1]
number_of_instances = X.shape[0]

print("Број и имиња на класи:", number_of_classes, target_names) 
print("Број и имиња на атрибути:", number_of_features, feature_names)
print("Број на податочни примероци:", number_of_instances)
```

 ### Визуализација на податоците
 
Во оваа подсекција ќе разгледаме неколку начини на визуализација на податоците со користење на хистограми, дијаграми на растурање и топлинска мапа на корелациска матрица.

#### Бокс-плот

Најпрво правиме таканаречен бокс-плот (boxplot), кој е достапен да се повика преку податочната рамка, и со него одеднаш се визуелизира распределбата на податоците за сите променливи. Со моделот се исцртува кутија која се протега од првиот $Q_1$ до третиот $Q_3$ квартил на податоците, а зелената линија ја означува медијаната $Q_2$. Дополнително се исцртуваат и така наречени "whiskers" (вид на тактилни мустаќи кај цицачите), кои имаат должина од 1.5 пати од меѓуквартилен опсег (inter-quartle range $IQR=Q_3-Q_1$). Со нив се опфaќаат податоците кои лежат еден IQR под $Q_1$ и над $Q_3$. Потоа податочните вредности кои припаѓаат надвор и од тој простор се означени со кругчиња и тие се нарекуваат нетипични вредности (outliers). Во некои случаи овие примероци би можело да се отстранат, доколку се увиди дека тие се случиле поради некоја грешка. Сепак, треба да се биде внимателен зашто понекогаш тие може да се одраз на некои ретки појави кои не би требало да ги отфрламе.

```python
fig = plt.figure(figsize=(9, 3.5))
df.boxplot()
plt.show()
```

Истиот гарфик може да го исцртам и за примероците кои припаѓаат во различните класи, т.е. според вредностите на целната променлива target. Воедно при овие визуелизации ја отфрламе целната променлива зашто е иста за сите примероци. Може да се забележи дека на овие графици кутиите се многу потесни, бидејќи има помала варијанса во податоците. Оттука, се детектираат и повеќе нетипични вредности.

```python
fig = plt.figure(figsize=(7, 3.5))
df[df.target==0].drop(['target'], axis=1).boxplot()
plt.title('Бокс-плот за примероците од класа setosa')
plt.show()
```

```python
fig = plt.figure(figsize=(7, 3.5))
df[df.target==1].drop(['target'], axis=1).boxplot()
plt.title('Бокс-плот за примероците од класа versicolor')
plt.show()
```

#### Хистограм на распределбата на вредонстите на атрибутите

Најпрво ќе ја разгледаме распределбата на вредностите за атрибутите со користење на хистограм, кој се прикажува со помош на функцијата sns.histplot(). Со опцијата kde=True се прикажува и апроксимативна крива која ја отсликува распределбата на вредностите. Прикажани се распределбите поединечно за секоја класа, како и за сите класи севкупно. Веднаш може да се забележи дека примероците од setosa значително се разликуваат од другите две, посебно во димензиите на петал листовите. Поради тоа иако вредностите на сите атрибути имаат едномодална дистрибуција кога се гледаат посебно по класи, за атрибутите висина и ширина на петал листовите се добиваат двомодални дистрибуции кога се гледаат севкупно сите примероци од сите класи.

```python
colors = ['purple','blue','orange']
fig = plt.figure(figsize=(10, 10))
fig.subplots(nrows=2, ncols=2)
for feature in range(number_of_features): # се прикажува посебен график за секој атрибут
    ax = plt.subplot(2,2, feature+1)
    plt.title(feature_names[feature]) 
    sns.histplot(X[:,feature],color='black',label='all classees',kde=True,stat="count") # се прикажува хистограм на вредностите на соодветниот атрибут без разлика во која класа припаѓаат
    for Class in range(number_of_classes): # се прикажува хистограм на вредностите на соодветниот атрибут за секоја класа посебно
        sns.histplot(X[y == Class,feature], color=colors[Class], label=target_names[Class],kde=True,stat="count") # Со X[y==c,i] се зема атрибутот i од примероците кои припаѓаат во класа c
    plt.legend()
plt.show()
```

#### Дијаграми на растурање

Дијаграмот на растурање (анг. scatter plot) може да се искористи за да се прикажат примероците во дводимензионален или тродимензионален простор. Во нашиот пример примероците ќе бидат прикажани во дводимензионален простори дефинирани со сите можни парови од четирите атрибути, што резултира во 4x4=16 графика. Во општ случај може да имаме голем број на атрибути, па тогаш би требало најрпво да ги избереме најзначајните атрибути кои би можел ода ги користиме за ваква визуелизација. Како мерки за тоа може да се искористат Пирсоновиот коефициент на корелација или взаемната информација.

```python
fig = plt.figure(figsize=(10, 10),layout='constrained')
fig.subplots(nrows=number_of_features, ncols=number_of_features)
for feature_i in range(number_of_features):
    for feature_j in range(number_of_features):  
        ax = plt.subplot(number_of_features,number_of_features,number_of_features*feature_i + feature_j+1)
        if feature_i==feature_j:
            for color, i, target_name in zip(colors, [0, 1, 2], target_names):
                plt.hist(X[y == i, feature_i], alpha=.6, color=color, label=target_name)
            plt.title("Атрибут "+str(feature_i)+" (хистограм)")
            plt.legend()
        else:
            for color, i, target_name in zip(colors, [0, 1, 2], target_names):
                plt.scatter(X[y == i, feature_i], X[y == i, feature_j], alpha=.6, color=color, label=target_name)
            plt.title("Атрибут "+str(feature_i)+" x Атрибут "+str(feature_j))
        plt.xticks(())
        plt.yticks(())
plt.show()

```

Сличен дијаграм може да се добие и со користење на библиотеката seaborn, која нуди повеќе можности, на следниот начин:

```python
df.columns[feature_j]
```

```python
fig = plt.figure(figsize=(10, 10),layout='constrained')
fig.subplots(nrows=number_of_features, ncols=number_of_features)
for feature_i in range(number_of_features):
    for feature_j in range(number_of_features):  
        ax = plt.subplot(number_of_features,number_of_features,number_of_features*feature_i + feature_j+1)
        if feature_i==feature_j:
            for color, i, target_name in zip(colors, [0, 1, 2], target_names):
                sns.histplot(X[y == i,feature_i], color=color, label=target_name,kde=True,legend=True) # Со X[y==c,i] се зема атрибутот i од примероците кои припаѓаат во класа c
        else:
            for color, i, target_name in zip(colors, [0, 1, 2], target_names):
                sns.scatterplot(data=df, x=df.columns[feature_j],y=df.columns[feature_i],hue='target',alpha=.6, palette=colors,legend=False)
        plt.xticks(())
        plt.yticks(())
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, target_names, loc='lower center',bbox_to_anchor=(.5, 1), ncol=3,frameon=False)
plt.show()
```

Алтернативно може да се користи и функцијата sns.pairplot(), која овозможува со една линија код да се прикажат дијаграмите на растурањe помеѓу сите променливи.

```python
g=sns.pairplot(df, hue="target", diag_kind="hist",palette=colors);
sns.move_legend(g, "lower center",bbox_to_anchor=(.5, 1),ncol=3, title=None);
for t, l in zip(g._legend.texts, target_names):
    t.set_text(l)
```

Истата функција 'pairplot' овозможува и исцртување на други типови на графици, како што е 'kernel density estimation (KDE)', со која се исцртуваат естимираните густини на распределба на вредностите на променливите.

```python
g=sns.pairplot(df, hue="target",kind="kde",palette=colors);
sns.move_legend(g, "lower center",bbox_to_anchor=(.5, 1),ncol=3, title=None);
for t, l in zip(g._legend.texts, target_names):
    t.set_text(l)
```

Дополнително може да се користи и функцијата seaborn.pairgrid(), која овозможува многу флексибилен начин за приказ на зависностата помеѓу сите парови атрибути, така што може да се комбинираат повеќе различни типови на графици во еден грид.

#### Мерки за зависности помеѓу атрибутите

На претходните графици можеше визуелно да се увиди кои атрибути се поврзани помеѓу себе. Сепак, за тоа постојат прецизно дефинирани мерки кои може да ги изразат зависностите квантитативно, како што се Пирсоновиот коефициент на корелација и взаемната информација. Тука ќе ја разгледаме примената само на Пирсоновиот коефициент на корелација.

Најпрво пресметуваме и визуелизираме корелациска матрица во која се претставени Пирсоновите коефициенти на корелација за сите парови атрибути и целната класа. Првите четири четири редици и колони соодветствуваат на атрибутите, а последните на класата.

```python
correlation_matrix = np.zeros((number_of_features+1,number_of_features+1))
# прво ги пресметуваме присоновите корелациски коефициенти за сите парови атрибути
for feature_i in range(number_of_features): 
    for feature_j in range(number_of_features):
        corr, _ = stats.pearsonr(X[:,feature_i], X[:,feature_j])
        correlation_matrix[feature_i][feature_j] = corr
# потоа ги пресметуваме пирсоновите корелациски коефициенти за секој атрибут со целната променлива
for feature_i in range(number_of_features): 
    target_i = number_of_features
    corr, _ = stats.pearsonr(X[:,feature_i], y)
    correlation_matrix[feature_i][target_i] = corr
    correlation_matrix[target_i][feature_i] = corr
# на крај го пресметуваме пирсоновиот корелациски коефициент за целната променлива само со себе (ова може и да се изостави зашто резултатот е 1)
corr, _ = stats.pearsonr(y, y)
correlation_matrix[target_i][target_i] = corr 
corr, _ = stats.pearsonr(X[:,feature_i], y)

plt.figure()
plt.imshow(correlation_matrix)
plt.xlabel("атрибут/класа")
plt.ylabel("атрибут/класа")
plt.colorbar()
plt.show()
```

```python
from scipy import linalg
from matplotlib import colors
import matplotlib as mpl

def plot_data(lda, X, y, y_pred):
    splot = plt.subplot(1, 1, 1) 
    
    true = y == y_pred  # Точно предвидени класи
    true0, true1, true2 = true[y == 0], true[y == 1], true[y == 2]
    X0, X1, X2 = X[y == 0], X[y == 1], X[y == 2]
    X0_true, X0_false = X0[true0], X0[~true0]
    X1_true, X1_false = X1[true1], X1[~true1]
    X2_true, X2_false = X2[true2], X2[~true2]

    # Исцртување на примероците од сите класи
    plt.scatter(X0_true[:, 0], X0_true[:, 1], alpha=.6, color="purple",label='setosa')
    plt.scatter(X0_false[:, 0], X0_false[:, 1], marker="x", s=20, color="purple")

    plt.scatter(X1_true[:, 0], X1_true[:, 1],  alpha=.6, color="blue",label='versicolor')
    plt.scatter(X1_false[:, 0], X1_false[:, 1], marker="x", s=20, color="blue")
    
    plt.scatter(X2_true[:, 0], X2_true[:, 1],  alpha=.6, color="orange",label='virginica')
    plt.scatter(X2_false[:, 0], X2_false[:, 1], marker="x", s=20, color="orange")

    # Исцртување на границите на одлука
    nx, ny = 100, 100
    x_min, x_max = plt.xlim()
    y_min, y_max = plt.ylim()
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx), np.linspace(y_min, y_max, ny))
    Z = lda.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    Z = Z[:, 1].reshape(xx.shape)
    plt.contour(xx, yy, Z, [0.5],colors="black")

    # Исцртување на средните вредности за сеекоја од класите
    plt.plot(lda.means_[0][0],lda.means_[0][1],"s",color="purple",markersize=15)
    plt.plot(lda.means_[1][0],lda.means_[1][1],"s",color="blue",markersize=15)
    plt.plot(lda.means_[2][0],lda.means_[2][1],"s",color="orange",markersize=15)
    return splot


def plot_ellipse(splot, mean, cov, color):
    v, w = linalg.eigh(cov)
    u = w[0] / linalg.norm(w[0])
    angle = np.arctan(u[1] / u[0])
    angle = 180 * angle / np.pi
    ell = mpl.patches.Ellipse(mean,2*v[0]**0.5,2*v[1]**0.5,angle=180+angle,facecolor=color)
    ell.set_clip_box(splot.axes.bbox)
    ell.set_alpha(0.2)
    splot.add_artist(ell)
    splot.set_xticks(())
    splot.set_yticks(())

def plot_lda_cov(lda, splot):
    plot_ellipse(splot, lda.means_[0], lda.covariance_, "purple")
    plot_ellipse(splot, lda.means_[1], lda.covariance_, "blue")
    plot_ellipse(splot, lda.means_[2], lda.covariance_, "orange")


def plot_qda_cov(qda, splot):
    plot_ellipse(splot, qda.means_[0], qda.covariance_[0], "purple")
    plot_ellipse(splot, qda.means_[1], qda.covariance_[1], "blue")
    plot_ellipse(splot, qda.means_[2], qda.covariance_[2], "orange")
```

Во следниот код е дадена визуелизација на LDA модел. Со квадрати се претставени средните вредности за примероците од секоја од класите $\mu_jc$, во редуцираниот дводимензионален простор, додека со елипсите се претставени двојните вредности од стандардните девијации $\sigma_jc$. Може да се забележи дека елипсите имаат иста форма, бидејќи коваријансните матрици се споделени помеѓу различните класи. Исто така границите на одлука помеѓу класите се прави линии, што е карактеристично за LDA. Од сликата може да се забележи дека примероците од класата 'setosa' се јасно оделени од примероците од другите класи. Од друга страна, примероците од другите две класи се наоѓаат поблиску во просторот, па два примерока од класата 'versicolor' се погрешно класифицирани како 'virginica', додека еден примерок од класата 'virginica' е погрешно класифициран како 'versicolor'.

```python
plt.figure(figsize=(8, 6))
plt.suptitle("Линеарна дискриминативна анализа")

lda = LinearDiscriminantAnalysis(solver="svd", store_covariance=True)
y_pred = lda.fit(X_reduced, y).predict(X_reduced)
splot = plot_data(lda, X_reduced, y, y_pred)
plot_lda_cov(lda, splot)
plt.axis("tight")
plt.legend()
plt.xlabel("димензија 1")
plt.ylabel("димензија 2")
plt.tight_layout()
plt.show()
```

Исто така некои од границите на одлука помеѓу класите, како што е таа помеѓу примероците од класите 'versicolo' и 'virginica' е квадратна функција, што е карактеристично за QDA. Од сликата може да се забележи дека примероците од класата 'setosa' повторно се јасно оделени од примероците од другите класи. Примероците од другите две класи и тука се наоѓаат поблиску во просторот, па дури три примерока од класата 'versicolor' се погрешно класифицирани како 'virginica', додека еден примерок од класата 'virginica' е погрешно класифициран како 'versicolor'.
