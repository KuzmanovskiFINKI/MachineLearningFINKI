## 1. Import all libraries





```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

## 2. Import the dataset (glass classification)

```python
dataset_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data"
columns = ['Id', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']
df = pd.read_csv(dataset_url, names=columns, index_col=0)
```

## 3. Preprocessing and initial information

```python
from itertools import count
y = df['Type']
X = df.drop(columns=['Type'])

feature_names = X.columns.values
target_names = list(set(y))
num_of_features = len(feature_names)
num_of_classes = len(target_names)
num_of_instances = len(X)

print("Број и имиња на класи:", num_of_classes, target_names)
print("Број и имиња на атрибути:", num_of_features, feature_names)
print("Број на податочни примероци:", num_of_instances)
```

## 5. Visualize the data

```python
# Boxplot (Q1,Q2-greenline,Q3,1.5IQR-whiskers...outliers)
fig = plt.figure(figsize=(10,50))
df.boxplot()
plt.show()
```

```python
# Histogram
colors = ['purple','blue','orange','red','yellow','green']
fig = plt.figure(figsize=(20, 20))
fig.subplots(nrows=3, ncols=3)
for feature in range(num_of_features): # се прикажува посебен график за секој␣атрибут
    ax = plt.subplot(3,3, feature+1)
    plt.title(feature_names[feature])
    sns.histplot(X.values[:,feature],color='black',label='all classees',kde=True,stat="count") # се прикажува хистограм на вредностите на соодветниот атрибут без разлика во␣која класа припаѓаат
    for Class in range(num_of_classes): # се прикажува хистограм на вредностите на␣соодветниот атрибут за секоја класа посебно
      sns.histplot(X.values[y == Class+1,feature], color=colors[Class], label=target_names[Class],kde=True,stat="count") # Со X[y==c,i] се зема атрибутот i од␣примероците кои припаѓаат во класа c
    plt.legend()
plt.show()
```

```python
# Scatter Plot (sns.pairplot()) --> Data shown in 2D/3D Space
g=sns.pairplot(df, hue="Type", diag_kind="hist",palette=colors);
sns.move_legend(g, "lower center",bbox_to_anchor=(.5, 1),ncol=3, title=None);
for t, l in zip(g._legend.texts, target_names):
  t.set_text(l)
```

```python
# Scatter Plot (sns.pairplot()) --> Data shown in 2D/3D Space
# KDE - estimated PDF (probability density function)
g=sns.pairplot(df, hue="Type", kind="kde",palette=colors);
sns.move_legend(g, "lower center",bbox_to_anchor=(.5, 1),ncol=3, title=None);
for t, l in zip(g._legend.texts, target_names):
  t.set_text(l)
```

```python
# Dependencies between features
# Correlation matrix

correlation_matrix = np.zeros((num_of_features+1,num_of_features+1))
# прво ги пресметуваме присоновите корелациски коефициенти за сите парови атрибути
for feature_i in range(num_of_features):
  for feature_j in range(num_of_features):
    corr, _ = stats.pearsonr(X.values[:,feature_i], X.values[:,feature_j])
    correlation_matrix[feature_i][feature_j] = corr
# потоа ги пресметуваме пирсоновите корелациски коефициенти за секој атрибут со␣целната променлива
for feature_i in range(num_of_features):
  target_i = num_of_features
  corr, _ = stats.pearsonr(X.values[:,feature_i], y)
  correlation_matrix[feature_i][target_i] = corr
  correlation_matrix[target_i][feature_i] = corr
# на крај го пресметуваме пирсоновиот корелациски коефициент за целната променлива␣само со себе (ова може и да се изостави зашто резултатот е 1)
corr, _ = stats.pearsonr(y, y)
correlation_matrix[target_i][target_i] = corr
corr, _ = stats.pearsonr(X.values[:,feature_i], y)

plt.figure()
plt.imshow(correlation_matrix)
plt.xlabel("атрибут/класа")
plt.ylabel("атрибут/класа")
plt.colorbar()
plt.show()
```

```python
# Lowering the dimension using LDA(Linear Discriminant Analysis)
LDA_reduce = LinearDiscriminantAnalysis(n_components=2) # Го поставуваме бројот на компоненти во новиот простор на 2
LDA_reduce.fit(X.values, y) # Го фитуваме LDA моделот на податоците X и y
X_reduced = LDA_reduce.transform(X.values) # Со помош на фитуваниот модел ги трансформираме влезните податоци X во но новите X_LDA

# Сега можеме да го визуелизираме податочното множество на еден дводимензионален график
plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2, 3, 4, 5], target_names):
  plt.scatter(X_reduced[y == i, 0], X_reduced[y == i, 1], alpha=.6, color=color, label=target_name)
plt.legend()
plt.xlabel("димензија 1")
plt.ylabel("димензија 2")
plt.title('Намалување на димензионалноста со LDA на Ирис податочното множество')
plt.show()
```

```python
# Reduced dimension KDE(Kernel Density Estimation) plot
df['димензија 1']= pd.DataFrame(X_reduced[:,0])
df['димензија 2']= pd.DataFrame(X_reduced[:,1])
g=sns.kdeplot(data=df, x="димензија 1", y="димензија 2",hue="Type",palette=colors, alpha=0.3,fill=True)

handles, labels = g.get_legend_handles_labels()
fig.legend(handles, target_names, labelcolor=colors, loc='lower center',bbox_to_anchor=(.5,1), ncol=3,frameon=False)
plt.show()
```

## 6. Classification [QDA, LDA, NaiveBayes]

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Models
QDA = QuadraticDiscriminantAnalysis()
LDA = LinearDiscriminantAnalysis()
NB = GaussianNB()

# Train the models
QDA.fit(X_train,y_train)
LDA.fit(X_train,y_train)
NB.fit(X_train,y_train)
# Check the priors
print("Приорните веројатности на класите користејќи QDA се:",QDA.priors_)
print("Приорните веројатности на класите користејќи LDA се:",LDA.priors_)
```

```python
# Test the models
y_pred = QDA.predict(X_test)
QDA_acc = accuracy_score(y_test,y_pred)

y_pred = LDA.predict(X_test)
LDA_acc = accuracy_score(y_test,y_pred)

y_pred = NB.predict(X_test)
NB_acc = accuracy_score(y_test,y_pred)

print("Точноста на класификацијата со методот QDA во оригиналното податочно множество е",round(QDA_acc,3)) # assumes that all the classes have different cov matrix, meaning that
print("Точноста на класификацијата со методот LDA во оригиналното податочно множество е",round(LDA_acc,3)) # suffers from when the classes are unbalanced, however we fixed this using the stratify=y when spliting the dataset | assumes all the classes have the same cov matrix
print("Точноста на класификацијата со методот NB во оригиналното податочно множество е",round(NB_acc,3)) # NB assumes that the features are independent, which are clearly not
```
